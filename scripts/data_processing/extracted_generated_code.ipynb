````python
import json
import os
import csv
import re
````

````python
def read_jsonl(file_path):
    data = []
    with open(file_path, 'r') as f:
        for line in f:
            data.append(json.loads(line))
    return data
````

````python
path_to_model_output = '../../results/mistral/model_output/mistral_codereval_python_v12.jsonl'
if not os.path.exists(path_to_model_output):
    raise Exception('Model output file does not exist')
````

````python
data = read_jsonl(path_to_model_output)
for item in data:
    response = item['complete_response']
    try:
        extracted_code = re.search(r'```python(.*?)```', response, re.DOTALL).group(1).strip()
    except:
        try:
            extracted_code = re.search(r'```(.*?)```', response, re.DOTALL).group(1).strip()
        except:
            print("Code could not be extracted")
            print(response)
            extracted_code = None
    item['generated_code'] = extracted_code

path_to_output = '../../results/mistral/model_output/mistral_codereval_python_v12_new.jsonl'
with open(path_to_output, 'w') as f:
    for item in data:
        f.write(json.dumps(item) + '\n')
````
